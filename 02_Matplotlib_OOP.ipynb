```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1, 20, 0.5)
y = x**2
x
```

```python
y
```

Versi matplotlib OOP

Versi 1

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul Grafik');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.54, 0.3, 0.3])

axes1.plot(x, y, color='red')
axes1.set_xlabel('Sumbu X besar')
axes1.set_ylabel('Sumbu Y besar')
axes1.set_title('Judul Grafik Utama')


axes2.plot(y, x, color='blue')
axes2.set_xlabel('Sumbu X')
axes2.set_ylabel('Sumbu Y')
axes2.set_title('Judul Grafik Kecil')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
fig.tight_layout()

# Plot for the first axes
axes[0].plot(x, y, color='red')
axes[0].set_xlabel('sumbu x1')
axes[0].set_ylabel('sumbu y1')
axes[0].set_title('Grafik 1')

# Plot for the second axes
axes[1].plot(y, x, color='green')
axes[1].set_xlabel('sumbu x2')
axes[1].set_ylabel('sumbu y2')
axes[1].set_title('Grafik 2')

fig.tight_layout()
```
