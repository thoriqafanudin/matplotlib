```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 20, 0.5)
y = x**2

# Mengatur ukuran dan dpi saat dicetak
fig = plt.figure(figsize=(4,4), dpi=100)

# Mengatur ukuran di jupyter
ax = fig.add_axes([0, 0, 0.5, 0.5]).scat(x, y)
```

```python
fig, axes = plt.subplots(2,1, figsize=(4,8), dpi=100)

axes[0].plot(x, y)
axes[0].set_title('Grafik 1')

axes[1].plot(x, y)
axes[1].set_title('Grafik 2')

plt.tight_layout()
plt.show()
```

```python
fig.savefig('gambarku.png')
```
