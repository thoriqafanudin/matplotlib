```python
import numpy as np
x=[10,20,30,40,50]
y=np.array(x)**2
plt.plot(x,y)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Judul Grafik')
plt.show()
```

```python
plt.subplot(131)
plt.plot(x,y,color='red')
plt.xlabel('Sumbu X1')
plt.ylabel('Sumbu Y1')
plt.title('Grafik 1')

plt.subplot(132)
plt.plot(y,x,color='blue')
plt.xlabel('Sumbu X2')
plt.ylabel('Sumbu Y2')
plt.title('Grafik 2')

plt.subplot(133)
plt.plot(y,x,color='green')
plt.xlabel('Sumbu X3')
plt.ylabel('Sumbu Y3')
plt.title('Grafik 3')

plt.tight_layout()
plt.show()
```
